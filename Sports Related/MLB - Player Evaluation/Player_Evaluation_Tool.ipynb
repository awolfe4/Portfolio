```python
import pandas as pd
```

```python
df = pd.read_csv('Amateur Batter Daataset.csv')
```

```python
# Function to convert percentage strings to float
def convert_percent(percentage):
    if isinstance(percentage, str):  # Check if it's a string
        return float(percentage.strip('%')) / 100
    return percentage  # Return the float as is

# List of columns to convert
percent_cols = [
    'AR K% ADJ', 'AR BB% ADJ', 'AR GB% ADJ', 'AR POP% ADJ',
    'Swing% AR', 'IZ Swing% AR', 'Chase% AR', 'Whiff% AR',
    'IZ Whiff% AR', 'OZ Whiff% AR'
]

# Apply conversion to the specified percentage columns
for col in percent_cols:
    df[col] = df[col].apply(convert_percent)
```

```python
# Step 1: Filter based on very relaxed criteria
filtered_df = df[
    (df['AR K% ADJ'] < 0.25) &                    # Low K% (25%)
    (df['AR BB% ADJ'] > 0.08) &                   # High BB% (8%)
    (df['Chase% AR'] < 0.25) &                    # Low Chase% (25%)
    (df['AR GB% ADJ'] < 0.45) &                   # Low GB% (45%)
    (df['AR POP% ADJ'] < 0.15) &                  # Low POP% (15%)
    (df['Whiff% AR'] < 0.25)                      # Low Whiff% (25%)
]
```

```python
# Step 1: Filter based on very relaxed criteria
filtered_df = filtered_df[
    (df['AR SLG ADJ'] > 0.390) &
    (df['AR ISO ADJ'] > 0.150) &
    (df['AR wOBA ADJ']> 0.310)
]
```

```python
# Step 3: Sort and select top players
top_players = filtered_df.sort_values(by='Conf Grade', ascending=False).head(5)
```

**Pitching Information**

```python
pitching_df = pd.read_csv('Pitching Data.csv')
#
```

```python
# Function to convert percentage strings to float
def convert_percent(percentage):
    if isinstance(percentage, str):  # Check if it's a string
        return float(percentage.strip('%')) / 100
    return percentage  # Return the float as is

# List of columns to convert
percent_cols = [
    'K%', 'BB%', 'GB%',
    'ADJ K%', 'ADJ BB%', 'ADJ GB%',
    'FPS%', 'Strike%', 'Zone%',
    'Chase%', 'Whiff%', 'zWhiff%', 'FB zWhiff%', 'xSpin Eff FB'
]

# Apply conversion to the specified percentage columns
for col in percent_cols:
    pitching_df[col] = pitching_df[col].apply(convert_percent)

columns_to_convert = ['xArm Angle FB', 'xArm Slot FB', 'Rel Speed FB', 'Spin Rate FB',
                      'Horz Break FB', 'Vert Break FB', 'Ext Above Exp FB',
                      'xSpin Eff FB', 'Spin Rate BB']

for col in columns_to_convert:
    pitching_df[col] = pd.to_numeric(pitching_df[col], errors='coerce')
```

```python
pitching_df.columns
```

```python
# Step 1: Filter based on easier criteria
filtered_pitchers = pitching_df[
    (pitching_df['ADJ K%'] > 0.15) &
     (pitching_df['ADJ BB%'] < 0.12) &
     (pitching_df['ADJ GB%'] > 0.40) &
      (pitching_df['ADJ SLG'] < 0.550)                 # Average ADJ SLG = 0.579
]
```

```python
filtered_pitchers = filtered_pitchers[
    (filtered_pitchers['FPS%'] > 0.60) &
    (filtered_pitchers['Whiff%'] > 0.13) &
    (filtered_pitchers['Strike%'] > 0.61)
]                    # High first pitch strike
```

```python
avg_rel_speed = filtered_pitchers['Rel Speed Max FB'].mean()
avg_xspin_eff = filtered_pitchers['xSpin Eff FB'].mean()
avg_ext_above_exp = filtered_pitchers['Ext Above Exp FB'].mean()

# Step 2: Create new columns for each player's standing relative to the average
filtered_pitchers['Rel Speed Max FB vs Avg'] = filtered_pitchers['Rel Speed Max FB'] - avg_rel_speed
filtered_pitchers['xSpin Eff FB vs Avg'] = filtered_pitchers['xSpin Eff FB'] - avg_xspin_eff
filtered_pitchers['Ext Above Exp FB vs Avg'] = filtered_pitchers['Ext Above Exp FB'] - avg_ext_above_exp
filtered_pitchers['Sum_of_Mechanic_Stats_in_Relation_to_the_Average'] = (filtered_pitchers['Rel Speed Max FB vs Avg'] +
                                                filtered_pitchers['xSpin Eff FB vs Avg'] +
                                                filtered_pitchers['Ext Above Exp FB vs Avg'])
```

```python
filtered_pitchers = filtered_pitchers.sort_values(
    by=['Rel Speed Max FB', 'xSpin Eff FB', 'Ext Above Exp FB','Sum_of_Mechanic_Stats_in_Relation_to_the_Average'],
    ascending=[False, False, False,False]  # All three columns sorted in descending order
)
```

```python
# Select only the desired columns
desired_columns = ['Player', 'Rel Speed Max FB', 'xSpin Eff FB', 'Ext Above Exp FB','Sum_of_Mechanic_Stats_in_Relation_to_the_Average']
filtered_pitchers = filtered_pitchers[desired_columns]
```
